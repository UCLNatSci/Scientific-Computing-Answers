```python
import matplotlib.pyplot as plt

pop = 100
r = 1.1
n = 10

pop_list = [pop]

for i in range(100):
    if i % 5 == 0:
        pop = pop - 31.7
    else:
        pop = pop * r
        pop_list.append(pop)

print(pop_list)

plt.plot(pop_list)
```

NB There are many ways to answer this question. The following only uses techniques described in the notes.

```python

x = [18, 19, 10, 19, 11, 7, 12, 6, 5, 4]
n = len(x)
x1 = []
for i in range(0, n, 2):
    x1.append(x[i])
    
print(x1)

x2 = []
for z in x:
    if z % 2 == 0:
        x2.append(z)
print(x2)
        
x3 = []
for i in range(n):
    x3.append(x[n-i-1])
print(x3)

x4 = [x[0], x[n-1]]
print(x4)
```
